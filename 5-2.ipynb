# 교차검증과 그리드 서치

```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine_csv_data')
data = wine[['alcohol',	'sugar', 'pH']]
target = wine[['class']]

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target)
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(train_input, train_target)
```

```python
sub_input.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(sub_input, sub_target)
print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
```

```python
dt.score(test_input, test_target)
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(dt, train_input, train_target) # 섞어서 폴드를 나누지는 않는다. 이미 tts로 섞었으므로..
print(scores)
```

- 1/5씩 사용해서 5번 테스트 했음을 알 수 있다(5-폴드 교차검증)

### 하이퍼파라미터
- 하이퍼파라미터를 찾아주고 위에서 cross_validate함수로 했던 교차검증을 해주는 클래스도 있다.

```python
from sklearn.model_selection import GridSearchCV

params = { # 테스트하고 싶은 파라미터를 그냥 다 집어넣으면 된다
    'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004]
}
gs = GridSearchCV(dt, params, n_jobs=-1)
```

```python
gs.fit(train_input, train_target)
```

```python
gs.best_estimator_ # 가장 점수가 잘 나오는 하이퍼 파라미터 값=뭐뭐뭐 입니다
```
