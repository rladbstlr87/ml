## iris
- 붓꽃(iris) 데이터를 이용해서 학습/검증 해보자

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris['species'].value_counts()
```

이제 `petal_length`,`petal_width` 특성기준으로 `setosa`인지 `versicolor` 인지 구별해보자

```python
cond = iris['species'].isin(['setosa', 'versicolor'])
lw_df = iris.loc[cond, ['petal_length', 'petal_width']]
flower_data = lw_df.values.tolist()
flower_target = [0] * 50 + [1] * 50
test_input = flower_data[60:]
test_target = flower_target[60:]
```

```python
# setosa = 0
# versicolor = 1
input_arr = np.array(flower_data)
target_arr = np.array(flower_target)
index = np.arange(100)
```

```python
np.random.shuffle(index)

train_input = input_arr[index[:60]]
train_target = target_arr[index[:60]]
test_input = input_arr[index[60:]]
test_target = target_arr[index[60:]]
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
```
